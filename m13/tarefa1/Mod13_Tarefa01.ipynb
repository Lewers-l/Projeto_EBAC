```python
renda.drop(axis =1, columns=['Unnamed: 0', 'data_ref', 'id_cliente'], inplace=True)
```

```python
a,b= patsy.dmatrices('np.log(renda) ~ C(sexo)+C(posse_de_veiculo)+C(posse_de_imovel)+'
                      'tipo_renda+educacao+estado_civil+tipo_residencia+'
                      'idade+tempo_emprego+qt_pessoas_residencia', data = renda)
```

```python
reg1 = sm.OLS(a,b).fit()
```

```python
reg1.summary()
```

```python
#sexo[M] - sendo o interceptor, por ter um p value considerável
c,d = patsy.dmatrices('np.log(renda) ~ C(sexo,Treatment(1))+C(posse_de_veiculo)+C(posse_de_imovel)+'
                      'tipo_renda+educacao+estado_civil+tipo_residencia+'
                      'idade+tempo_emprego+qt_pessoas_residencia', data = renda)
```

```python
reg2 = sm.OLS(c,d).fit()
```

```python
reg2.summary()
```

```python
# com isso, vimos que educacao é a variável que pode ser excluida
```

```python
e,f = patsy.dmatrices('np.log(renda) ~ C(sexo,Treatment(1))+C(posse_de_veiculo)+C(posse_de_imovel)+'
                      'tipo_renda+estado_civil+tipo_residencia+'
                      'idade+tempo_emprego+qt_pessoas_residencia', data = renda)
```

```python
reg3 = sm.OLS(e,f).fit()
```

```python
reg3.summary()
```

```python
y,x = patsy.dmatrices('np.log(renda) ~ C(sexo,Treatment(1))+C(posse_de_veiculo)+C(posse_de_imovel)+'
                      'idade+tempo_emprego+qt_pessoas_residencia', data = renda)
```

```python
reg4 = sm.OLS(y,x).fit()
```
