```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df = pd.read_csv('esse.csv')
df = df.head(50)
df1 = df[['APGAR1','APGAR5','SEXO','GESTACAO','ESTCIVMAE']]
fig , axes = plt.subplots(2,3)
plt.rc('figure', figsize=(20, 20))
_ = sns.countplot(ax = axes[0 , 0], x='APGAR1', data=df1)
_ = sns.countplot(ax = axes[0 , 1], x='APGAR5', data=df1)
_ = sns.countplot(ax = axes[0 , 2], x='SEXO', data=df1)
a4 = sns.countplot(ax = axes[1 , 0], x='GESTACAO', data=df1)
_ = sns.countplot(ax = axes[1 , 1], x='ESTCIVMAE', data=df1)
labels = a4.set_xticklabels(a4.get_xticklabels(),rotation=30)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close("all")
df1['soma'] = df1['APGAR5'].sum()
df1['media'] = df1['APGAR5']/df1['soma']
plt.figure(figsize=(5, 5))
sns.barplot(x='APGAR5',y='media', data=df1)
```
